```python
N = len(data)
mean = sum(data)/N
diff = data.ptp()
disp = sum((data - mean)**2)/N
std = np.sqrt(disp)
# std = (std**2+0.7**2)**0.5
# y = np.exp(-(data-mean)**2/(2*std**2))/(2*np.pi)/std
x = np.linspace(mean - std*3, mean + std*3, 100)
y = np.exp(-(x - mean)**2/(2*std**2))/(2*np.pi)**0.5/std
```

```python
def draw(bins):
    plt.close()
    # plt.title(f"Гистограмма для $m={bins}$")
    plt.hist(data, bins=bins, histtype='step', color='k', density=True)
    plt.plot(x, y, c='k', label=f"$\\langle R \\rangle = {mean:.1f}\\ Ом$\n$\\sigma = {std:0.1f}\\ Ом$")
    plt.legend(handlelength=0)
    plt.xlabel("$R, Ом$")
    plt.ylabel("$w$")
    plt.savefig(f"hist{bins}.svg")

draw(10)
draw(15)
draw(20)
```
